## Task 2: Core Libraries

Run the cell below to verify that all required libraries are installed. All four should show a version number and print **OK**.

````python
import importlib

required = {
    "numpy":       "np",
    "pandas":      "pd",
    "matplotlib":  "mpl",
    "sklearn":     "sklearn"
}

all_ok = True
for package, alias in required.items():
    try:
        mod = importlib.import_module(package)
        version = getattr(mod, "__version__", "installed")
        print(f"{package:15s} {version:15s} OK")
    except ImportError:
        print(f"{package:15s} {'MISSING':15s} FAIL")
        all_ok = False

print()
if all_ok:
    print("All core libraries installed.")
else:
    print("Some libraries are missing. See CE49X_Setup_Guide.ipynb for installation instructions.")
````

## Task 3: Quick Smoke Test

Run the cell below. If everything is set up correctly, it will produce a small plot. This confirms NumPy, Pandas, and Matplotlib all work together.

````python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample civil engineering data
np.random.seed(42)
cement_content = np.array([250, 300, 350, 400, 450])  # kg/m³
compressive_strength = 0.08 * cement_content + np.random.normal(0, 2, 5)  # MPa

fig, ax = plt.subplots(figsize=(6, 3.5))
ax.scatter(cement_content, compressive_strength, color='steelblue', s=80, zorder=3)
ax.set_xlabel('Cement Content (kg/m³)')
ax.set_ylabel('Compressive Strength (MPa)')
ax.set_title('Concrete Mix — Smoke Test')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Smoke test passed — NumPy, Pandas, and Matplotlib are working.")
````

Run the cell below from **inside your cloned repository folder**.

````python
import subprocess

def run(cmd):
    result = subprocess.run(cmd, shell=True, capture_output=True, text=True)
    return result.stdout.strip() if result.returncode == 0 else f"ERROR: {result.stderr.strip()}"

print(f"Git version:   {run('git --version')}")
print(f"Git user.name: {run('git config user.name')}")
print(f"Git user.email:{run('git config user.email')}")
print()
print("Remotes:")
print(run('git remote -v'))
````
